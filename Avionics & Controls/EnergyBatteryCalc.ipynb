```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
thrust = [.486, 1.952,4.410, 7.878, 12.384, 17.965, 24.670, 32.559, 41.704, 52.163, 63.801, 74.719]
torque = [13.708, 9.953, 7.003, 5.277, 4.078, 3.089, 2.255, 1.564, 1.006, .574, .263, .070]
torque = torque[::-1]


cs = CubicSpline(thrust, torque)
xs = np.arange(0, 75, 0.1)

plt.plot(thrust, torque, 'o', label='data')

plt.plot(xs, cs(xs), label="S")

calctorque = cs(6.25)

print(calctorque)

plt.show()
```
